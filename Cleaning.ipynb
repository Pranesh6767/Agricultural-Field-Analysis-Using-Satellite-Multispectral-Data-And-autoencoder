```python
rootpath = '2_data_structured'
dest = '3_data_cleaned'
for i in datasheets:
    filepath = os.path.join(rootpath,i)
    despath = os.path.join(dest,i)
    df = pd.read_csv(filepath)
    df = df.dropna(axis=0)
    df = df.astype('str')
    columns = df.columns.tolist()
    for j in columns[1:]:
        df[j] = df[j].str.replace(',', '').astype(float)
    nir = df['B5']
    red = df['B4']
    df['ndvi'] =(nir-red)/(nir+red)
    el1 = (nir*2)+1
    el1 = el1.pow(2)
    el2 = nir - red
    el2 = 8 * el2
    el3 = el1 - el2
    el3 = el3.pow(1/2)
    el4 = (nir*2)+1
    el5 = el4 - el3
    el5 = el5/2
    df['msavi2'] = el5
    ndvi_mean = np.mean(df['ndvi'])
    msavi2_mean = np.mean(df['msavi2'])
    ndvi_max = df['ndvi'].max()
    msavi2_max = df['msavi2'].max()
    if (df.isnull().any().any()==False) and ((ndvi_max<1) and (ndvi_max>-1)) and ((msavi2_max<1) and (msavi2_max>-1)):
        df.to_csv(despath)
    else:
        print('Faulty Entry found....!    Filename:',str(i))
    
```
