# Slicing the data

### In this part we are selecting some specific number of rows (satellite sensor data collected at different time instance) from each datapoint
### From the data analysis we found that the slicing threshold should be 39
### So we will first sort out the rows in each datapoint based on average ndvi index....
### After that we will select 39 rows from each of the datapoint
### After that we will flatten the dataset 
### And after flattening we will append all the datapoints in single dataframe

```python
import pandas as pd
import numpy as np
import os
```

# Testing on single sheet

```python
rootdir = '3_data_cleaned'
datasheets = os.listdir(rootdir)
i = 0
filename = os.path.join(rootdir,datasheets[i])
df = pd.read_csv(filename)
df.head()
```

```python
rootdir = '3_data_cleaned'
dest = '5_data_processed'
datasheets = os.listdir(rootdir)

for i in datasheets:
    filename = os.path.join(rootdir,i)
    df = pd.read_csv(filename)
    df.sort_values(by=['ndvi'], inplace=True, ascending=False)
    sliced_df = df.iloc[:39,2:]
    desname = os.path.join(dest,i)
    sliced_df.to_csv(desname)
```

```python
def create_data_point(filename):
    df = pd.read_csv(filename)
    sliced_df = df.iloc[:,1:-2]
    df_np = np.array(sliced_df)
    df_np = df_np.flatten()
    ndvi_mean = np.mean(df['ndvi'])
    msavi2_mean = np.mean(df['msavi2'])
    df_np = np.append(df_np,[ndvi_mean,msavi2_mean])
    return df_np
```

```python
test_feature = create_data_point('5_data_processed/0.csv')
no_of_features = test_feature.shape[0]
```

```python
no_of_features
```

```python
main_df = pd.DataFrame()
for j in range(no_of_features):
    main_df[str(j)] = None
```

```python
main_df.shape
```

```python
rootpath = '5_data_processed'
datasheets = os.listdir(rootpath)
for i in datasheets:
    filepath = os.path.join(rootpath,i)
    feature_vector = create_data_point(filepath)
    main_df.loc[len(main_df)] = feature_vector
```
